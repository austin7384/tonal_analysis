```python
import pandas as pd
```

```python
df = pd.read_csv('/data/processed/llm_evaluated/clean_evaluations/clean_test_prompt1_run1.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)
df1 = pd.read_csv('/data/processed/llm_evaluated/clean_evaluations/clean_test_prompt1_run1.csv')
df1.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df = pd.merge(df, df1, how='left', on='ArticleID')
print(list(df.columns))
df.head()
```

```python
RUBRICS = [
    "Modal Verb Strength",
    "Hedging Frequency & Type",
    "Qualifier Density",
    "Acknowledgement of Limitations",
    "Caution-Signaling Connectors",
    "Assertiveness & Voice",
    "Active/Passive Voice Ratio",
    "Sentence Length & Directness",
    "Imperative-Form Occurrence",
    "Pronoun Commitment",
    "Novelty-Claim Strength",
    "Jargon/Technicality Density",
    "Emotional Valence",
    "Evidence & Citation Usage",
    "Practical/Impact Orientation",
    "Readability",
]

avg_abs_diff = {}

for rubric in RUBRICS:
    avg_abs_diff[rubric] = (
        (df[f"{rubric}_x"] - df[f"{rubric}_y"])
        .abs()
        .mean()
    )

avg_abs_diff_df = pd.DataFrame.from_dict(
    avg_abs_diff, orient="index", columns=["Mean Absolute Difference"]
)

print(avg_abs_diff_df)
```
